## **Análise Temporal**

```python
#Para fazer uma análise das datas eu vou criar algumas variáveis usando datas. 
#Primeiro vou converter a coluna data_cliente da tabela base para tipo date

base['data_cliente'] = pd.to_datetime(base['data_cliente'])
base['data_aprovacao'] = pd.to_datetime(base['data_aprovacao'])
#Agora vou criar outras variáveis com informações como Mês, ano, dia, nome do dia.


base['Ano'] = base['data_aprovacao'].dt.year
base['Mes'] = base['data_aprovacao'].dt.strftime('%b')
base['Dia'] = base['data_aprovacao'].dt.day
base['Semana_ano'] = base['data_aprovacao'].dt.isocalendar().week
base['Trimestre'] = base['data_aprovacao'].dt.quarter
base['Dias_ate_chegar'] = ( base['data_cliente'] -  base['data_aprovacao']).dt.days
```

```python
#Traduzindo os nomes dos meses para português

base['Mes'] = base['Mes'].map({'Jan':'jan',
                                   'Feb':'fev',
                                   'Mar':'mar',
                                   'Apr':'abr',
                                   'May':'maio',
                                   'Jun':'jun',
                                   'Jul':'jul',
                                   'Aug':'ago',
                                   'Sep':'set',
                                   'Oct':'out',
                                   'Nov':'nov',
                                   'Dec':'dez',})
```

### *Quantidade de pedidos*

```python
lista = ['jan','fev','mar','abr','maio','jun','jul','ago','set','out','nov','dez']
base[base['Ano'] == 2017].groupby(['Mes'])['id_ordem'].count().to_frame().reindex(lista).reset_index()
```

```python
def barra_contagem_plotly(ano,dados):
  lista = ['jan','fev','mar','abr','maio','jun','jul','ago','set','out','nov','dez']
  dados = dados[dados['Ano'] == ano].groupby(['Mes'])['id_ordem'].count().to_frame().reindex(lista).reset_index()
  fig = px.bar(dados, x='Mes', y='id_ordem',text_auto='.2s',category_orders=lista)
  fig.show()
```

```python
barra_contagem_plotly(2016,base)
barra_contagem_plotly(2017,base)
barra_contagem_plotly(2018,base)
```

```python
def linha_contagem_plotly(ano,dados):
  lista = ['jan','fev','mar','abr','maio','jun','jul','ago','set','out','nov','dez']
  dados = dados[dados['Ano'] == ano].groupby(['Mes'])['id_ordem'].count().to_frame().reindex(lista).reset_index()
  dados = dados.fillna(0)
  fig = px.line(dados, x='Mes', y='id_ordem',category_orders=lista)
  fig.show()
```

```python
linha_contagem_plotly(2016,base)
linha_contagem_plotly(2017,base)
linha_contagem_plotly(2018,base)
```

### *Quantidade de itens*

```python
def barra_soma(ano,dados,y):
    lista = ['jan','fev','mar','abr','maio','jun','jul','ago','set','out','nov','dez']
    dados = dados[dados['Ano'] == ano].groupby(['Mes'])[y].sum().to_frame().reindex(lista).reset_index()
    fig = px.bar(dados, x='Mes', y=y,text_auto='.2s')
    fig.show()
```

```python
barra_soma(2016,base,'numero_de_itens')
barra_soma(2017,base,'numero_de_itens')
barra_soma(2018,base,'numero_de_itens')
```

```python
def linha_soma(ano,dados,y):
    lista = ['jan','fev','mar','abr','maio','jun','jul','ago','set','out','nov','dez']
    dados = dados[dados['Ano'] == ano].groupby(['Mes'])[y].sum().to_frame().reindex(lista).reset_index()
    dados = dados.fillna(0)
    fig = px.line(dados, x='Mes', y=y)
    fig.show()
```

```python
linha_soma(2016,base,'numero_de_itens')
linha_soma(2017,base,'numero_de_itens')
linha_soma(2018,base,'numero_de_itens')
```

## *Criando perfil do cliente por estado*

```python
def perfil(estado):
    print('Itens mais comprados')
    print(base[base['estado_cliente'] == estado].groupby(['categoria_produto'])['numero_de_itens'].sum().sort_values(ascending=False).head(6))
    print('\n')
    
    print('Média no valor do frete')
    print(base[base['estado_cliente'] == estado]['frete'].mean())
    print('\n')
    
    print('Média de dias de espera')
    print(base[base['estado_cliente'] == estado]['Dias_ate_chegar'].mean())
    print('\n')
```

```python
perfil('SP')
```

```python
perfil('RJ')
```

```python
perfil('MG')
```

```python
perfil('RS')
```

```python
perfil('PR')
```

### Conclusão
- Entre os estados que mais compram não há uma grande variação nos tipos de itens vendidos. Oque mais varia é a média no valor do frete e na espera sendo uma variação de 6 reais no valor do frete e 13 dias no tempo de espera.
